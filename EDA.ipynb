#### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Loading the dataset

```python
pd.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('./data/raw/DSL-StrongPasswordData.csv')
dataset.head()
```

```python
dataset.isna().sum()
```

```python
i=1
plt.figure(figsize=(24,24))
for column in dataset.columns[3:]:
    plt.subplot(6,6,i)
    plt.hist(dataset[column])
    plt.title(column+' distribution')
    plt.xlabel('values')
    plt.ylabel('frequencies')
    i+=1
plt.show()
```
